# Connecting to a Database

```python
import psycopg2

# Database connection parameters
host = "localhost"
port = "5432"
dbname = "cos101_db"
user = "postgres"
password = "cos101"

# Establish the connection
try:
    connection = psycopg2.connect(
        host=host,
        port=port,
        dbname=dbname,
        user=user,
        password=password
    )
    print("Connection to PostgreSQL DB successful")

    # Create a cursor object
    cursor = connection.cursor()

    # Execute a SQL query
    cursor.execute("SELECT version();")
    
    # Fetch and print the result of the query
    db_version = cursor.fetchone()
    print(f"PostgreSQL database version: {db_version}")

    # Close the cursor and connection
    cursor.close()
    connection.close()

except Exception as error:
    print(f"Error connecting to PostgreSQL DB: {error}")
```

# Selecting existing Database

```python
import psycopg2

# Database connection parameters
host = "localhost"
port = "5432"
dbname = "cos101_db"
user = "postgres"
password = "cos101"


try:
    connection = psycopg2.connect(
        host=host,
        port=port,
        dbname=dbname,
        user=user,
        password=password
    )
    print("Connection to PostgreSQL DB successful")

    # Create a cursor object
    cursor = connection.cursor()

    # Define the SQL SELECT query
    select_query = "SELECT * FROM company;"

    # Execute the query
    cursor.execute(select_query)

    # Fetch all rows from the executed query
    data = cursor.fetchall()

    print("column", " | ", "column2")
    print("------------------------")
        
    # Iterate through the rows and print them
    for row in data:
        print(f"{row[0]}       |  {row[1]}")

    # Close the cursor and connection
    cursor.close()
    connection.close()

except Exception as error:
    print(f"Error connecting to PostgreSQL DB: {error}")
```

# Create a new Database

```python
import psycopg2
from psycopg2 import sql

# Database connection parameters
host = "localhost"
port = "5432"
dbname = "postgres"
user = "postgres"
password = "cos101"


# Name of the new database to be created
new_dbname = "cos102_db"

try:
    # Connect to the PostgreSQL server
    connection = psycopg2.connect(
        host=host,
        port=port,
        dbname=dbname,
        user=user,
        password=password
    )
    connection.autocommit = True  # Enable autocommit mode
    print("Connection to PostgreSQL server successful")

    # Create a cursor object
    cursor = connection.cursor()

    # Define the SQL statement to create a new database
    create_db_query = sql.SQL("CREATE DATABASE {}").format(sql.Identifier(new_dbname))

    # Execute the SQL statement
    cursor.execute(create_db_query)
    print(f"Database '{new_dbname}' created successfully")

    # Close the cursor and connection
    cursor.close()
    connection.close()

except Exception as error:
    print(f"Error creating the database: {error}")
```

# Create a Table

```python
import psycopg2


# Database connection parameters
host = "localhost"
port = "5432"
dbname = "cos102_db"
user = "postgres"
password = "cos101"


try:
    # Connect to the PostgreSQL database
    connection = psycopg2.connect(
        host=host,
        port=port,
        dbname=dbname,
        user=user,
        password=password
    )
    connection.autocommit = True  # Enable autocommit mode
    
    print("Connection to PostgreSQL DB successful")

    # Create a cursor object
    cursor = connection.cursor()

    # Define the SQL statement to create a new table
    create_table_query = """
    CREATE TABLE sst_info (
        id SERIAL PRIMARY KEY,
        name VARCHAR(50) NOT NULL,
        age INT NOT NULL,
        department VARCHAR NOT NULL,
        password VARCHAR NOT NULL
    );
    """

    # Execute the SQL statement
    cursor.execute(create_table_query)
    print("Table created successfully")

    # Close the cursor and connection
    cursor.close()
    connection.close()

except Exception as error:
    print(f"Error creating table: {error}")
```

# INSERT into Table

```python
import psycopg2

# Database connection parameters
host = "localhost"
port = "5432"
dbname = "cos102_db"
user = "postgres"
password = "cos101"


try:
    # Connect to the PostgreSQL database
    connection = psycopg2.connect(
        host=host,
        port=port,
        dbname=dbname,
        user=user,
        password=password
    )
    connection.autocommit = True  # Enable autocommit mode
    print("Connection to PostgreSQL DB successful")

    # Create a cursor object
    cursor = connection.cursor()

    # Define the SQL INSERT statement
    insert_query = """
    INSERT INTO SST_INFO (id, name, age, department, password)
    VALUES (%s,%s,%s,%s,%s);
    """

    
    # Data to be inserted
    data_to_insert = (1, "Jamel Ojo", 20, "Software Engineering", "anthelo4")

    # Execute the SQL statement
    cursor.execute(insert_query,data_to_insert)
    print("Data inserted successfully")

    # Close the cursor and connection
    cursor.close()
    connection.close()

except Exception as error:
    print(f"Error inserting data: {error}")
```

# SELECT from Table

```python
import psycopg2

# Define your database connection parameters
db_params = {
    'dbname': 'cos102_db',
    'user': 'postgres',
    'password': 'cos101',
    'host': 'localhost',
    'port': '5432'  # Default is '5432'
}

try:
    # Establish a connection to the database
    connection = psycopg2.connect(**db_params)

    # Create a cursor object
    cursor = connection.cursor()

    # Define the SQL query
    select_query = "SELECT * FROM sst_info WHERE age < 30"

    # Execute the SQL query
    cursor.execute(select_query)

    # Fetch all the results
    result = cursor.fetchall()

    # Print the results
    for row in result:
        print(row)

except Exception as error:
    print(f"Error while connecting to PostgreSQL: {error}")

finally:
    # Close the cursor and connection to clean up
    if cursor:
        cursor.close()
    if connection:
        connection.close()
```

```python
import psycopg2
from psycopg2 import sql
import tkinter as tk
db_params = {
    'dbname': 'postgres',
    'user': 'postgres',
    'password': 'cos101',
    'host': 'localhost',
    'port': '5432'
}


        
class Database:
    def __init__(self,db_params,db_name = "_"):
        self.db_name = db_name
        self.db_params = db_params 
        try:
            self.__update_connection() 
        except Exception as error:
            print(f"Error whilst connect to Postgre: {error}")
    def __close_connection(self):
        self.cursor.close()
        self.connection.close()
    
    def __update_connection(self,db_name = 'postgres'):
            self.db_params['dbname'] = db_name
            self.connection = psycopg2.connect(**db_params)
            self.cursor = self.connection.cursor()
            self.connection.autocommit = True      
        
    def create_database(self):
        self.__update_connection()
        try:
            create_db_query = sql.SQL("CREATE DATABASE {}").format(sql.Identifier(self.db_name))
            self.cursor.execute(create_db_query)
            print(f"Database '{self.db_name}' created successfully")
        except Exception as error:
            print(f"Error Whilst Creating Database: {error}")
                
    def create_table(self,table_name):
        self.__update_connection(self.db_name)
        try:
            command = f"""
            CREATE TABLE {table_name}(
            name VARCHAR PRIMARY KEY,
            division VARCHAR NOT NULL,
            age INT NOT NULL,
            module INT NOT NULL,
            constraint chk_MODULE CHECK(module IN (1,2,3,4,5)),
            constraint chk_division CHECK(division IN ('BUSINESS', 'ENGINEERING', 'PRODUCT', 'DESIGN', 'COMMERCIAL' , 'REVENUE'))
            )
            """
            self.cursor.execute(command)
            print(f"Table {table_name} succesfully created")
        except Exception as error:
            print(f"Error Whilst creating database: {error}")
            
    def add_data(self,data,table):
        self.__update_connection(self.db_name)
        try:
            insert_command = f"""
            INSERT INTO {table} (name,division,age,module)
            VALUES (%s,%s,%s,%s)
            """
            self.cursor.execute(insert_command,data)
            print("Data Inserted Successfully")
            self.__close_connection()
        except Exception as error:
            print(f"Failed to append data: {error}")
            
    def query_data(self,query,table,db = ''):
        if db == '':
            db = self.dbname
        self.__update_connection(db)
        try:
            self.cursor.execute(query)
            result = self.cursor.fetchall()
            for row in result:
                print(row)
            return result
        except Exception as error:
            print(f"Failed to Get query: {error}")
            
    def list_databases(self):
        self.__update_connection()
        self.cursor.execute("SELECT datname FROM pg_database WHERE datistemplate = false;")
        databases = [row[0] for row in self.cursor.fetchall()]
        return databases
        self.__close_connection()
        
    def list_tables(self,database):
        self.__update_connection(database)
        try:
            self.cursor.execute("SELECT current_database() as DBName;")
            print(f"Database {self.cursor.fetchall()}")
            
            command = f"""
            SELECT table_name 
            FROM information_schema.tables 
            WHERE table_schema = 'public';
            """
            self.cursor.execute(command,(database,))
            table_data = [ row[0] for row in self.cursor.fetchall() ]
            return table_data
        except Exception as error:
            print(f"Failed to Get field Information: {error}")
        
    def list_fields(self,table,database):
        self.__update_connection(database)
        try:
            self.cursor.execute("SELECT current_database() as DBName;")
            print(f"Database {self.cursor.fetchall()}")
            
            command = f"""
            SELECT column_name, data_type
            FROM information_schema.columns
            WHERE table_name = '{table}'
            AND table_schema = 'public';
            """
            self.cursor.execute(command,(table,))
            field_data = [ row[0] for row in self.cursor.fetchall() ]
            return field_data
        except Exception as error:
            print(f"Failed to Get field Information: {error}")
        
databases = Database(db_params).list_databases()
field_name = Database(db_params).list_fields('test_table', 'test_db')
tables = Database(db_params).list_tables('test_db')
print(tables)
print(field_name)
```

```python
import tkinter as tk
from tkinter import *

class GUI:
    def __init__(self, Database_object):
        self.database_object = a
        self.target = None

    def main_window(self):       
        self.main_window = tk.Tk()
        self.main_window.title('Database Manager')
        self.main_window.geometry('500x200')

        tk.Label(self.main_window, text="Select an Action").pack()
        tk.Button(self.main_window, text="Query a Database", command=self.database_query).pack()
        tk.Button(self.main_window, text="Modify a Database", command=self.Login).pack()

        self.main_window.mainloop()

    def Login(self):
        pass
        
    def condition_insert(self, previous_window,table,database):
        table = table.get()
        if previous_window:
            previous_window.withdraw()
        self.new_window = tk.Toplevel(previous_window)
        fields = self.database_object.list_fields(table,database)
        entries = []
        entry_values = []
        
        for index in range(len(fields)):
            _ = tk.Label(self.new_window, text = fields[index]).pack()
            entry = tk.Entry(self.new_window)
            entry.pack()
            entries.append(entry)
            
        def command_run(previous_window,queries,table,database):
            if previous_window:
                previous_window.withdraw()
            where_argument = ""
            for query in queries:
                if query != '':
                    if where_argument != '':
                        where_argument += ' AND '
                    where_argument += query
            command = f"""
            SELECT *
            FROM {table}
            WHERE {where_argument};
            """
            x = self.database_object.query_data(command,table,database)
            print(x)
            
        def getter():
            for entry in entries:
                entry_values.append(entry.get().upper())
            print(entry_values)
            command_run(self.new_window,entry_values,table,database)
            entry_values.clear()
            
        _ = tk.Button(self.new_window, text = "Get Query", command = getter).pack()
            
        
        
    def table_query(self):
        target = self.target.get()
        options = self.database_object.list_tables(f"{target}")
        self.run(self.new_window,"Select a table to Query",lambda :self.condition_insert(self.new_window,self.target,target),options)
        
    def database_query(self):
        databases = self.database_object.list_databases()
        self.run(self.main_window,"Select Database to Query", self.table_query,databases)
        
    def run(self,previous_window,instruction,submit_function,options):
        if previous_window:
            previous_window.withdraw()  
        self.new_window = tk.Toplevel(previous_window)
        tk.Label(self.new_window, text=instruction).pack()
        
        self.target = StringVar(self.new_window)
        menu = OptionMenu(self.new_window,self.target, *options)
        menu.pack()
        _ = tk.Button(self.new_window, text = 'Submit', command = submit_function).pack()
        
        

    def show_main_window(self):
        if self.querier_window:
            self.querier_window.destroy()  
        if self.main_window:
            self.main_window.deiconify()  
            
a = Database(db_params)
app = GUI(a)
app.main_window()
```
